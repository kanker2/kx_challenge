```python
import torch
from torch.quantization import quantize_dynamic, default_qconfig
from sentence_transformers import SentenceTransformer

import pandas as pd
import pykx as kx

logs = pd.read_csv('logfile.csv')
model = SentenceTransformer("paraphrase-MiniLM-L3-V2",device='cpu')
```

```python
torch.set_num_threads(3)
tqint8 = torch.qint8

qmodel = quantize_dynamic(model, {torch.nn.Linear}, dtype=tqint8)

def embed(text):
    return qmodel.encode(text)

logs['embeddings'] = logs.text.apply(embed).apply(lambda x: " ".join(x.astype(str).tolist()))
logs.to_csv('logfile_embedded.csv', index=False)

qry_embedded_in_q_format = " ".join(embed("rescue search mission").astype(str).tolist())
print(qry_embedded_in_q_format)
```
